```python
columnas_numericas = data.select_dtypes(include=['float', 'int'])

supdata = data[columnas_numericas.columns]


supdata = supdata.drop(["RowNumber", "CustomerId"], axis=1)
supdata
```

```python
def titanic_corr(data): #matriz de correlación
    correlation = data.corr()
        
    plt.figure(figsize=(12, 10))

    sns.heatmap(correlation, annot=True, cbar=True, cmap="RdYlGn")

titanic_corr(supdata)
```
